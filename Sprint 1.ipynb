```python
import pandas as pd
```

```python
data = pd.read_csv('datos.csv')
```

```python
def ordenar_dataframe(data, columna, ascendente=True):
    return data.sort_values(by=columna, ascending=ascendente)
```

```python
columna = input("Ingrese el nombre de la columna por la cual desea ordenar: ")
tipo_orden = input("Ingrese el tipo de ordenamiento (ascendente o descendente): ")

ascendente = True if tipo_orden.lower() == "ascendente" else False

data_ordenada = ordenar_dataframe(data, columna, ascendente)

display(data_ordenada)
```
